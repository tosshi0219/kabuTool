```python
from kabuMemo.fetcher import fetcher as memo_fetcher
from kabuTool.calculator.data import Data
memofet = memo_fetcher()
kabuka = memofet.fetch_kabuka(3496)

data = Data(kabuka)
```

```python
import matplotlib.pyplot as plt

df = data.annual_data
max_drawdown_result = data.calc_max_drawdown()
max_increase_result = data.calc_max_increase()

print(max_drawdown_result)
plt.figure(figsize=(20,3))
plt.plot(df.date, df.close)
plt.axvline(max_drawdown_result['start_date'])
plt.axvline(max_drawdown_result['end_date'])
plt.show()

print(max_increase_result)
plt.figure(figsize=(20,3))
plt.plot(df.date, df.close)
plt.axvline(max_increase_result['start_date'])
plt.axvline(max_increase_result['end_date'])
plt.show()

target_date = '2020-11-22'
print(data.calc_year_increase(target_date))
```
